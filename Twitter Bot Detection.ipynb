```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("account_type", axis = 1)
X = pd.get_dummies(X,drop_first=True)
y = df["account_type"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
LogisticModel = LogisticRegression()
solver = ['multinomial','lbfgs','sag', 'saga', 'newton-cg', 'liblinear']
penalty = [None, "l1", "l2", "elasticnet"]
C = [0.001, 0.01, 0.1, 0.5, 1, 5, 10, 100]
param_grid = {"solver":solver, "penalty":penalty, "C":C}
```

```python
grid = GridSearchCV(LogisticModel, param_grid, n_jobs = -1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
LogisticModel = LogisticRegression(C = 5, penalty = "l2", solver = 'lbfgs')
LogisticModel.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
n_neighbors = list(range(1,30,2))
metric = ['euclidean', 'manhattan', 'minkowski']
weights = ['uniform', 'distance']
param_grid = {"n_neighbors":n_neighbors, "metric":metric, "weights":weights}
```

```python
grid = GridSearchCV(knn, param_grid, verbose = 3, n_jobs = -1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
knn = KNeighborsClassifier(metric = "manhattan", n_neighbors = 23, weights = "distance")
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_model = GradientBoostingClassifier(n_estimators = 1000)
learning_rate= [0.1, 0.5, 1]
subsample = [0.5, 0.7, 1]
max_depth = [3,5,7]
param_grid = {"learning_rate":learning_rate, "subsample":subsample, "max_depth":max_depth}
param_grid_two = {"subsample":subsample, "max_depth":max_depth}
grid = GridSearchCV(gb_model, param_grid, cv = 3, n_jobs = -1, verbose = 3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
gb_model = GradientBoostingClassifier(learning_rate = 0.1, n_estimators = 1000, max_depth = 7, subsample = 0.7)
# {'learning_rate': 0.1, 'n_estimators': 500}
gb_model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier()
max_features = ["sqrt", "log2"]
n_estimators = [10, 100, 1000]
param_grid = {"max_features":max_features, "n_estimators":n_estimators}
```

```python
grid = GridSearchCV(random_forest, param_grid, cv = 5, verbose = 3, n_jobs = -1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
random_forest = RandomForestClassifier(max_features = "sqrt", n_estimators = 1000)
random_forest.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay
```

```python
def metrics(model):
    pred = model.predict(X_test)
    print(classification_report(y_test, pred))
    ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
    RocCurveDisplay.from_estimator(model,X_test,y_test)
```

```python
metrics(LogisticModel)
```

```python
metrics(knn)
```

```python
metrics(random_forest)
```

```python
metrics(gb_model)
```

```python
print("Based on the information, the gradient boosted model provides the best representation of the data.")
print("Both the random forest and the gradient boosted model perform the same, but the gradient boosted model is better since")
print("more bots are correctly labeled as bots for the gradient boosted and the main goal is to correctly label bots.")
print("The AUC score for both the gradient boosted model and the random forest model is the same".)
```
