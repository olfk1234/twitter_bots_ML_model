```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("account_type", axis = 1)
X = pd.get_dummies(X,drop_first=True)
y = df["account_type"]
X_train, X_obj, y_train, y_obj = train_test_split(X, y, test_size=0.3, random_state=101)
X_eval, X_test, y_eval, y_test = train_test_split(X_obj, y_obj, test_size=0.5, random_state=101)
```

```python
X_eval
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
LogisticModel = LogisticRegression(random_state = 101)
solver = ['multinomial','lbfgs','sag', 'saga', 'newton-cg', 'liblinear']
penalty = [None, "l1", "l2", "elasticnet"]
C = [0.001, 0.01, 0.1, 0.5, 1, 5, 10, 100]
param_grid = {"solver":solver, "penalty":penalty, "C":C}
```

```python
grid = GridSearchCV(LogisticModel, param_grid, n_jobs = -1)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
LogisticModel = LogisticRegression(C = 100, penalty = "l2", solver = 'newton-cg', random_state = 101)
LogisticModel.fit(X_train, y_train)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_model = GradientBoostingClassifier(n_estimators = 1000, random_state = 101)
learning_rate= [0.1, 0.5, 1]
subsample = [0.5, 0.7, 1]
max_depth = [3,5,7]
param_grid = {"learning_rate":learning_rate, "subsample":subsample, "max_depth":max_depth}
param_grid_two = {"subsample":subsample, "max_depth":max_depth}
grid = GridSearchCV(gb_model, param_grid, cv = 3, n_jobs = -1, verbose = 3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
gb_model = GradientBoostingClassifier(learning_rate = 0.1, n_estimators = 1000, max_depth = 7, subsample = 0.7, random_state = 101)
# {'learning_rate': 0.1, 'n_estimators': 500}
gb_model.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(random_state = 101)
max_features = ["sqrt", "log2"]
n_estimators = [10, 100, 1000]
param_grid = {"max_features":max_features, "n_estimators":n_estimators}
```

```python
grid = GridSearchCV(random_forest, param_grid, cv = 5, verbose = 3, n_jobs = -1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
random_forest = RandomForestClassifier(max_features = "sqrt", n_estimators = 1000, random_state = 101)
random_forest.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay
```

```python
def metrics(model):
    pred = model.predict(X_eval)
    print(classification_report(y_eval, pred))
    ConfusionMatrixDisplay.from_estimator(model, X_eval, y_eval)
    RocCurveDisplay.from_estimator(model,X_eval,y_eval)
```

```python
metrics(LogisticModel)
```

```python
metrics(random_forest)
```

```python
metrics(gb_model)
```
